# Principal Component Analysis

Why PCA?

  . Visualisation

  . Dimensionality Reduction

  . Multicollinearity
  
  . Noise reduction in the dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Iris (1).csv')
df.head()
```

```python
labels=df["Species"]
data=df.drop(["Id","Species"],axis=1)
```

```python
print(data.shape)
print(labels.shape)
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
standardize_data=StandardScaler().fit_transform(data)
print(standardize_data.shape)
```

```python
standardize_data[:5]
```

# PCA using SCIKIT-LEARN

```python
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(standardize_data)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
np.arange(len(np.cumsum(pca.explained_variance_ratio_)))
```

**OBSERVATION:** As we can see that only 2 PCA components are explaining 95.8% variance, so lets go ahead and consider these two components.

```python
# configuring the parameteres
# number of components = 2
pca_new=PCA(n_components=2)
pca_new_data=pca_new.fit_transform(standardize_data)

print('shape=',pca_new_data.shape)
```

```python
pca_df=pd.DataFrame(data=pca_new_data,columns=['1st_principal','2nd_principal'])
pca_df['Labels']=labels.values
pca_df.head()
```
